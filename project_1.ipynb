```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= df.drop("Unnamed: 0",axis = 1)
print(df)
```

```python
# gender destributions 

plt.figure(figsize=(5,5))
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# From the above chart analysis :
# The number of females in the data is more the the number of males. 
```

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(gb,annot= True)
plt.show()
```

```python
# From the above chat we have calculated that the education of the parentes have a good impact .
```
